```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(rc = {"figure.figsize":(16, 9)})
```

```python
out = np.loadtxt( "out.txt")

net = np.linspace(0, 1, len(out))
net2 = np.linspace(0, 1, 1000)
plt.subplot(211)
plt.scatter(net, out, label="  ")
plt.plot(net2, - net2 * (net2 - 2) / 2, label=r" : $y = \frac{-x(x-2)}{2}$")
plt.title(r" , 4 ,  $-y'' = 1, \quad y(0) = y'(1) = 0$")
plt.legend()
plt.subplot(212)
plt.scatter(net, out + net * (net - 2) / 2, label="    ", color="r")
plt.title(r"  ")
plt.legend()
# plt.plot(net2, - net2 * (net2 - 1) / 2, label=r" : $y = \frac{x(x-1)}{2}$")
plt.savefig("out.pdf")
np.max(np.fabs(- net * (net - 2) / 2 - out))
```

```python
out = np.loadtxt( "out.txt")

def y(x):
    C = - 1 / ((np.exp(1))**2 + 1)
    return C * np.exp(x) + C * np.exp(1) * np.exp(1) * np.exp(- x) + 1

net = np.linspace(0, 1, len(out))
plt.subplot(211)
plt.scatter(net, out, label="  ")
plt.plot(net, y(net), label=r" :  $y = C e^x + C e^2 e^{-t} + 1$,  $C = -\frac{1}{ 1 + e^2}$")
plt.title(r" , 4 ,  $-y'' + y = 1, \quad y(0) = y'(1) = 0$")
plt.legend()
plt.subplot(212)
plt.scatter(net, out - y(net), label="    ", color="r")
plt.title(r"  ")
plt.legend()
plt.savefig("out.pdf")
np.max(np.fabs(y(net) - out))
```
